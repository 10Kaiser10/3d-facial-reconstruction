Two stereo images undLeft and undRight are loaded. These images are rectified versions of the initial images. Rectification process was done in the **stereoRectification.ipynb** notebook (code not documented in that notebook, so might be difficult to understand). Also images as cropped to make them square.

```python
import cv2
import matplotlib.pyplot as plt

left = cv2.imread("../inputImages/undLeft.png")
#left = cv2.cvtColor(left, cv2.COLOR_BGR2RGB)[:,512:3552]
left = left[:,512:3552]

right = cv2.imread("../inputImages/undRight.png")
#right = cv2.cvtColor(right, cv2.COLOR_BGR2RGB)[:,512:3552]
right = right[:,512:3552]
```

# Generating mask and pyramid

```python
#threshold values to remove the grey background
a = 150
b = 200

#generating masks for the images
maskL = cv2.inRange(left, (a,a,a), (b,b,b))
maskL = cv2.bitwise_not(maskL)
maskR = cv2.inRange(right, (a,a,a), (b,b,b))
maskR = cv2.bitwise_not(maskR)
```

```python
#displaying the inital images, masks and masked image
fig, axs = plt.subplots(2,3,figsize=(12,9))
axs[0][0].imshow(left)
axs[0][1].imshow(maskL)
axs[0][2].imshow(cv2.bitwise_and(left, left, mask=maskL))
axs[1][0].imshow(right)
axs[1][1].imshow(maskR)
axs[1][2].imshow(cv2.bitwise_and(right, right, mask=maskR))
```

Generating image pyramid for both images. Although this pyramid is not used in the current implementation.

```python
leftPyr = [left]
rightPyr = [right]
for i in range(5):
  left = cv2.pyrDown(left, dstsize=(left.shape[0] // 2, left.shape[1] // 2))
  right = cv2.pyrDown(right, dstsize=(right.shape[0]//2, right.shape[1]//2))
  leftPyr.append(left)
  rightPyr.append(right)
```

# Pixel Matching Test

Testing out the pixel matching algorithm by matching the pixel along the epipolar line.

```python
#point to be matched
i=52
j=50

#taking the lowest resolution image from the pyramid
leftimg = leftPyr[-1].copy()
rightimg = rightPyr[-1].copy()

#temp is the template to be searched, res is result of matching along the epipolar line
temp = leftimg[i-1:i+2, j-1:j+2]
res = cv2.matchTemplate(rightimg[i-1:i+2,:], temp, cv2.TM_CCORR_NORMED)

#displaying results
recl = leftimg.copy()
cv2.rectangle(recl, (j-1,i-1), (j+2,i+2), (255,0,0), 1)
recr = rightimg.copy()

m = res.argmax()    #argmax of res gives the matched point
print('Matching pixel at:', m+1)
print('Disparity: ', j-m-1)

cv2.rectangle(recr, (m,i-1), (m+3,i+2), (255,0,0), 1)

fig, axs = plt.subplots(1,2,figsize=(10,5))
axs[0].imshow(recl)
axs[1].imshow(recr)
plt.show()
```

# Disparity Map

Generating a left and right disparity map. Also generating a matchLeft/Right which keeps y index of the matched pixel array for calculating uniqueness constraint.

```python
import numpy as np

#creating a copy of the low res images
leftimg = leftPyr[-1].copy()
rightimg = rightPyr[-1].copy()

#initializing the output arrays
dispMapLeft = np.zeros(leftimg.shape[:2])
dispMapRight = np.zeros(leftimg.shape[:2])
matchLeft = np.zeros(leftimg.shape[:2])
matchRight = np.zeros(rightimg.shape[:2])

#loop though all pixel (except the border), and match each pixel with the other image
for i in range(1, leftimg.shape[0]-1):
  for j in range(1, leftimg.shape[1]-1):
    temp = leftimg[i-1:i+2, j-1:j+2]
    res = cv2.matchTemplate(rightimg[i-1:i+2,:], temp, cv2.TM_CCORR_NORMED)
    matchLeft[i][j] = res.argmax()+1
    dispMapLeft[i][j] = j-res.argmax()-1

for i in range(1, rightimg.shape[0]-1):
  for j in range(1, rightimg.shape[1]-1):
    temp = rightimg[i-1:i+2, j-1:j+2]
    res = cv2.matchTemplate(leftimg[i-1:i+2,:], temp, cv2.TM_CCORR_NORMED)
    matchRight[i][j] = res.argmax()+1
    dispMapRight[i][j] = j-res.argmax()-1
```

```python
mapL = dispMapLeft.copy()
mapR = dispMapRight.copy()

#clip the extreame points in the map, just for visulization
mapL[mapL<-5] =-5
mapL[mapL> 25] = 25
mapR[mapR> 5] = 5
mapR[mapR<-25] =-25

fig,axs = plt.subplots(1,2,figsize=(10,20))
axs[0].imshow(mapL, cmap='gray')
axs[1].imshow(25-mapR, cmap='gray')
```

A lot of points are have weird matches, this can be cleaned up by applying the constraints,

# Constraints

```python
uniqueness = np.ndarray(dispMapLeft.shape, dtype='uint8')

#go though all points and check if point p is matched to q in other image, then q must be matched to p
for i in range(dispMapLeft.shape[0]):
  for j in range(dispMapLeft.shape[1]):
    if(matchRight[i, int(matchLeft[i, j])] == j):
      uniqueness[i][j] = 255
    else:
      uniqueness[i][j] = 0

#displaying the map, yellow means point satisfies constraint
plt.imshow(uniqueness)
```

```python
ordering = np.ndarray(dispMapLeft.shape, dtype='uint8')

#go though all points and make sure that disparity of point p does not exceed its right neighbour by more than one
for i in range(dispMapLeft.shape[0]):
  for j in range(dispMapLeft.shape[1]-1):
    if(dispMapLeft[i,j]-dispMapLeft[i,j+1] <= 1):
      ordering[i][j] = 255
    else:
      ordering[i][j] = 0

plt.imshow(ordering)
```

```python
smoothness = np.ndarray(dispMapLeft.shape, dtype='uint8')

#go though all points and make sure that all their 8 neighbour's disparity does not differ by more than 1
for i in range(1,dispMapLeft.shape[0]-1):
  for j in range(1,dispMapLeft.shape[1]-1):
    morethanone = 0
    for x in range(-1,2):
      for y in range(-1,2):
        if(abs(dispMapLeft[i,j]-dispMapLeft[i+x,j+y]) > 1):
          morethanone += 1
    if(morethanone <= 4):
      smoothness[i][j] = 255
    else:
      smoothness[i][j] = 0


plt.imshow(smoothness)
```

# Disparity map of points satisfying constraints

```python
#apply constraints, if not satisfied, set disparity to 0
faceDisp = cv2.bitwise_and(dispMapLeft, dispMapLeft, mask=uniqueness)
faceDisp = cv2.bitwise_and(faceDisp, faceDisp, mask=smoothness)
faceDisp = cv2.bitwise_and(faceDisp, faceDisp, mask=ordering)

#also apply mask to cut of the background and only display face
masknew = cv2.inRange(leftPyr[-1], (a,a,a), (b,b,b))
masknew = cv2.bitwise_not(masknew)
faceDisp = cv2.bitwise_and(faceDisp, faceDisp, mask=masknew)

plt.imshow(faceDisp)
plt.colorbar()
```

Still there are points with extreme values, these are mostly points which were mismatched, but satisfy constraints. Plotting histogram of all values:

```python
h = plt.hist(faceDisp.flatten(), bins=50, )
```

The histogram shows there are some points with extreme values outside the range 0-20. We need to rematch these points, also rematch points in the face which fail the constraints. Creating an array showing all points that need to be rematched:

```python
#this array shows points that need rematching, points must be inside the face, 
#and have current disparity outside range 5-25 to be remached
invalidDisp = np.zeros(faceDisp.shape, dtype='uint8')
invalidDisp[faceDisp <= 5] = 255
invalidDisp[faceDisp > 25] = 255
invalidDisp[masknew == 0] = 0
```

```python
#histogram of valid points i.e. valid point in the face and the background
h = plt.hist(faceDisp[invalidDisp==0].flatten(), bins=50, )
```

```python
#showing all points to be rematched
plt.imshow(invalidDisp)
```

# Rematching

```python
#creating array to store rectified disparity values after remaching
rectifiedDisp = faceDisp.copy()
rectifiedInvalid = invalidDisp.copy()
avgs = np.zeros(faceDisp.shape) #this array was just used for debugging

#loop through invalid points, find average disparity of valid neighbours, then use this disparity value to restrict search area
#then search again with the same method used above
for i in range(1, leftimg.shape[0]-1):
  for j in range(1, leftimg.shape[1]-1):
    if(invalidDisp[i,j]==255):
        
      #calculate average disparity in neighbourhood  
      sumDisp = 0
      validPix = 0
      for x in range(-1,2):
        for y in range(-1,2):
          sumDisp += rectifiedDisp[i+x,j+y]*(255-rectifiedInvalid[i+x,j+y])/255
          validPix += (255-rectifiedInvalid[i+x,j+y])/255
      
      #if no valid neighbours, then stop
      if (validPix == 0):
        continue
      
      #else, find matching pixel in a restricted search area
      avgDisp = int(sumDisp/validPix)
      avgs[i,j] = avgDisp
      temp = leftimg[i-1:i+2, j-1:j+2]
      res = cv2.matchTemplate(rightimg[i-1:i+2,j-avgDisp-3:j-avgDisp+4], temp, cv2.TM_CCORR_NORMED)
      rectifiedDisp[i][j] = avgDisp - res.argmax() + 2
      rectifiedInvalid[i][j] = 0
        
      #if rectified disparity is negative then mark point as invalid  
      if(rectifiedDisp[i][j]<=0):
        rectifiedDisp[i][j] = 0
        rectifiedInvalid[i][j] = 255
```

```python
#plot of invalid points after rematching
plt.imshow(rectifiedInvalid)
```

```python
#plot of rectified disparity values
plt.imshow(rectifiedDisp)
plt.colorbar()
```

Much cleaned plot, still some areas like forehead and mouth are not well matched, but good enough.

```python
#plotting histogram of new disparity values
h=plt.hist(rectifiedDisp.flatten(),20)
```

# Calculating disparity in high resolution image

The disparity values from the low res images will be used to restrict search space while matching the high res image

```python
bigImgL = leftPyr[0]
bigImgR = rightPyr[0]
bigDisp = np.zeros(bigImgL.shape[:2])
```

```python
#using same method as above for matching, this time only searching in a restricted area
#go through all unmasked points
for i in range(1, bigImgL.shape[0]-1):
  for j in range(1, bigImgL.shape[1]-1):
    if(maskL[i,j] == 0):
      continue
    
    #check disparity of corresponding pixel from small images 
    smallDisp = rectifiedDisp[i//32][j//32]
    
    #search in a 64 pixel long neighbourhood around the pixel corresponding to the match in the low res image
    temp = bigImgL[i-1:i+2, j-1:j+2]
    res = cv2.matchTemplate(bigImgR[i-1:i+2,int(j-32*smallDisp-32):int(j-32*smallDisp+32)], temp, cv2.TM_CCORR_NORMED)
    bigDisp[i][j] = 32*smallDisp - (res.argmax() + 1) + 32
```

```python
#plotting high res disparity values
plt.imshow(bigDisp)
plt.colorbar()
```

```python
#histogram of disparity
h = plt.hist(bigDisp.flatten(), 50)
```

# Depth Map

```python
#clipping extreame disparity values
disp = bigDisp.copy()
disp[bigDisp < 300] = 300
disp[bigDisp > 700] = 700
```

```python
#calculating depth by dividing (focal length * baseline distance) by disparity. Rough values of these parameters is used
depthTest = 0.35*4000/disp
plt.imshow(depthTest, cmap='binary')
plt.colorbar()
```

```python
#creating a map to store invalid disparity values, these are too high or low and wont be used in point cloud creation
invalid = np.zeros(bigDisp.shape)
invalid[bigDisp < 300] = 255
invalid[bigDisp > 700] = 255
plt.imshow(invalid)
```

```python
#final depth map
depth = np.zeros(bigDisp.shape)
depth[invalid == 0] = 4876.8*0.32/bigDisp[invalid==0]
```

# Point cloud

Creating point cloud by using x and y values of the pixel and depth values as z component. Not the most accurate method, but it works fine.

```python
#making a list of points to be plotted
x_points = []
y_points = []
z_points = []
colors = []

for i in range(0, bigImgL.shape[0]):
  for j in range(0, bigImgL.shape[1]):
    if (invalid[i][j] == 0):
      x_points.append(i)
      y_points.append(j)
      z_points.append(depth[i,j]*200)
      colors.append(bigImgL[i,j])

x_points = np.array(x_points)
y_points = np.array(y_points)
z_points = np.array(z_points)
c_array = np.array(colors)/255
```

```python
#randomly sampling 500000 points from the cloud beacuse the entire set of points is in millions which is too high.
index = np.random.choice(x_points.shape[0], 500000, replace=False)  
```

```python
%matplotlib notebook
from mpl_toolkits import mplot3d
fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection="3d")
ax.scatter(y_points[index], z_points[index], -1*x_points[index], s=0.01, c=c_array[index])
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
```
